# Seaborn Library:

```python
#Introduction: Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.
```

```python
#Import Matplotlib
import matplotlib.pyplot as plt

#Import Seaborn
import seaborn as sns
```

```python
#Plotting a Distplot
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([0, 1, 2, 3, 4, 5])
plt.show()
```

```python
#Plotting a Distplot Without the Histogra
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot([0, 1, 2, 3, 4, 5], hist=False)
plt.show()
```

```python
#Releational plots:
```

```python
#Scatter Plot: Visualizing the relationship between two variables
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Tip vs Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
#Line Plot:
import seaborn as sns
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
sns.lineplot(x=x, y=y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Joint plot:
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="scatter")
plt.show()
```

```python
#Categorical plots:
```

```python
#Bar Plot: Comparing categories using bars
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
#Box Plot: Displaying the distribution of a continuous variable across different categories
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
#Count plot:
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
sns.countplot(x="class", data=titanic)
plt.title("Passenger Class Count")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
#Distribution Plots:
```

```python
#Histogram: Showing the distribution of a single variable
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.show()
```

```python
#KDE plot:
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
sns.kdeplot(data=titanic["age"], shade=True)
plt.title("Kernel Density Estimate of Age")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
#Rug plot:
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.rugplot(x="total_bill", data=tips, height=0.3)
plt.title("Rug Plot of Total Bill")
plt.xlabel("Total Bill")
plt.yticks([])
plt.show()
```

```python
#Regression plots:
```

```python
#Linear regression plot:
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Linear Regression Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
#Residual plot:
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.residplot(x="total_bill", y="tip", data=tips, lowess=True, line_kws={'color': 'red'})
plt.title("Residual Plot")
plt.xlabel("Total Bill")
plt.ylabel("Residuals")
plt.show()
```

```python
#Clustermap:
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
titanic = titanic.dropna()
data = titanic[['survived', 'pclass', 'age', 'fare']]
sns.clustermap(data, cmap='coolwarm', standard_scale=1)
plt.title("Clustermap of Titanic Dataset")
plt.show()
```

```python
#Multi-plot Grids:
```

```python
#FaceGrid:
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
g = sns.FacetGrid(titanic, row="class", col="who")
g.map(plt.hist, "age")
g.fig.suptitle("Age Distribution by Class and Gender")
plt.tight_layout()
plt.show()
```

```python
#PairGrid:
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
iris = sns.load_dataset("iris")
g = sns.PairGrid(iris, hue="species")
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.add_legend()
plt.show()
```

```python
#JointGrid:
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot)
g.ax_marg_x.hist(tips["total_bill"], color="skyblue", alpha=0.6, bins=15)
g.ax_marg_y.hist(tips["tip"], color="salmon", alpha=0.6, bins=15, orientation="horizontal")
g.set_axis_labels(xlabel="Total Bill", ylabel="Tip")
plt.show()
```

```python
#Violin plot:
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```
